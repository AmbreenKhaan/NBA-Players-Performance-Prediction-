## **Data Preprocessing**

### Converting String to Numeric Variables

```python
# Convert 'draft_year' to numeric, coerce errors to NaN and replace NaNs with 0
nba_dataset['draft_year'] = pd.to_numeric(nba_dataset['draft_year'], errors='coerce').fillna(0).astype(int)

# Convert 'draft_round' to numeric, coerce errors to NaN and replace NaNs with 0
nba_dataset['draft_round'] = pd.to_numeric(nba_dataset['draft_round'], errors='coerce').fillna(0).astype(int)

# Convert 'draft_number' to numeric, coerce errors to NaN and replace NaNs with 0
nba_dataset['draft_number'] = pd.to_numeric(nba_dataset['draft_number'], errors='coerce').fillna(0).astype(int)

# Display the data types of all columns after the conversion
nba_dataset.dtypes
```

### Dropping Columns

```python
nba_dataset.drop(columns=(['Unnamed: 0', 'team_abbreviation', 'college', 'country', 'season']), axis=1, inplace=True, errors='ignore')

# Display the remaining column names to confirm the columns have been dropped
nba_dataset.columns
```

## **Exploratory Data Analysis (EDA)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize distributions
metrics = ['pts', 'reb', 'ast', 'net_rating']
for metric in metrics:
    sns.histplot(nba_dataset[metric], kde=True)
    plt.title(f'Distribution of {metric}')
    plt.show()

# Pairplot for key metrics
sns.pairplot(nba_dataset, vars=['pts', 'reb', 'ast', 'net_rating'])
plt.show()
```

## **Feature Engineering**

```python
# Create new features: rolling averages
nba_dataset['rolling_pts'] = nba_dataset['pts'].rolling(window=5).mean().fillna(nba_dataset['pts'])
nba_dataset['rolling_ast'] = nba_dataset['ast'].rolling(window=5).mean().fillna(nba_dataset['ast'])


print(nba_dataset.head())  # Preview changes
```

### Correlation Analysis

```python
correlation_matrix =nba_dataset.select_dtypes(exclude=['object']).corr()

# Plot the correlation matrix as a heatmap for better visualization
plt.figure(figsize=(16, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.4f', linewidths=0.5)  # Create the heatmap with annotations
plt.title('Correlation Matrix of Players Dataset')  # Set the title of the plot
plt.show()  # Display the plo
```

```python

# Get correlations with the target variable 'pts'
pts_correlation = correlation_matrix['pts']

print(pts_correlation)
```

## **Feature Selection**

```python
# Select columns based on the correlation thresholds:
selected_columns = pts_correlation[(pts_correlation > 0.2) | # Select features with a correlation greater than 0.20 with PTS
                                   (pts_correlation < -0.1)  # Select features with a correlation less than -0.10 with PTS
                                   ].index.tolist()

# Display the selected columns
selected_columns
```

### Feature and Label Selection

```python
# The columns that we will be making predictions with.
X = nba_dataset[selected_columns].drop(columns=['pts'])  # Select the feature columns, excluding 'pts' (target)

# The column that we want to predict.
Y = nba_dataset['pts']  # Set 'pts' as the target variable (points scored by players)
```

```python
X.info()
```

## Data Scaling

```python
scaler = StandardScaler()  # Standardize numerical features (mean=0, std=1)

# Transform numerical columns
X = scaler.fit_transform(X)  # Apply scaling to numerical data
```

## Splitting Data into training and testing datasets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=2)  
# 75% for training, 25% for testing
```

## **Model Training**

### Linear Regression 

```python
# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
# Making predictions on the test set  
y_pred_lr = linear_model.predict(X_test) # Predicting points (Y) for test data
```

### Ridge Model

```python
# Creating and training a Ridge Regression model with regularization
ridge_model = Ridge()  # Instantiating the Ridge model
ridge_model.fit(X_train, Y_train)  # Fitting the model on the training data

# Making predictions on the test set
Y_pred_ridge = ridge_model.predict(X_test)  # Predicting points (Y) for the test data
```

### Evaluation of Both Models

```python

# Calculate evaluation metrics for Linear Regression
r2_lr = r2_score(Y_test, y_pred_lr)
mae_lr = mean_absolute_error(Y_test, y_pred_lr)
mse_lr = mean_squared_error(Y_test, y_pred_lr)

# Calculate evaluation metrics for Ridge Regression
r2_ridge = r2_score(Y_test, Y_pred_ridge)
mae_ridge = mean_absolute_error(Y_test, Y_pred_ridge)
mse_ridge = mean_squared_error(Y_test, Y_pred_ridge)

# Create a DataFrame to present the results
results_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Ridge Regression'],
    'R2 Score': [r2_lr, r2_ridge],
    'Mean Absolute Error': [mae_lr, mae_ridge],
    'Mean Squared Error': [mse_lr, mse_ridge]
})

# Display the results
print(results_df)
```

## Feature Analysis and Conclusion

### Model Performance Comparison

The performance of **Linear Regression** and **Ridge Regression** is nearly identical, with only slight variations in the evaluation metrics. Both models exhibit almost the same values for **Mean Squared Error (MSE)**, **R-Squared (R²)**, and **Mean Absolute Error (MAE)**. This suggests that the regularization in **Ridge Regression** does not provide a substantial improvement over **Linear Regression** for this dataset. 

As a result, both models perform similarly, with **Linear Regression** emerging as a simpler and equally effective choice for predicting player performance. 

### Feature Contribution Summary

- **'pts' (Points):** Points scored directly measure player performance, making it the most significant feature in both models. It typically has a high positive coefficient, indicating that as this metric increases, player performance (in terms of points) also increases.
  
- **'ast' (Assists):** Assists indicate a player's playmaking ability, and while indirectly related to scoring, this feature positively impacts the model's prediction of performance.

- **'usg_pct' (Usage Percentage):** A higher **usage percentage** signifies more involvement in offensive plays, contributing positively to points scored and player performance.

- **'ts_pct' (True Shooting Percentage):** This measure of shooting efficiency is crucial for maximizing points. Players with higher shooting efficiency tend to score more, making **ts_pct** an important predictor of performance.

### Conclusion

While **Ridge Regression** includes regularization to handle multicollinearity and prevent overfitting, its impact on model performance is minimal compared to **Linear Regression** in this case. Both models deliver similar results, with **Linear Regression** offering a simpler and equally effective solution for predicting player performance.
